```python
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
from shapely.geometry import Point
from mpl_toolkits.basemap import Basemap
import matplotlib.cm as cm
import seaborn as sns
```

```python
# CSV dosyasını oku
df = pd.read_csv("deprem_verisi.csv")
```

```python
# Sadece >=6.0 büyüklükte olanları al
df = df[(df["Tarih"].notna()) & (df["Boylam"].notna()) & (df["xM"].notna()) & (df["xM"] >= 6.0)]
```

# 1980 öncesi verileri kullanarak model eğitme

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# 1) 1980 öncesi veri ve model
df_before_1980 = df[df["Yil"] < 1980].dropna(subset=["Yil", "Boylam"]).copy()
X_train = df_before_1980[["Yil"]].values
y_train = df_before_1980["Boylam"].values

model = LinearRegression().fit(X_train, y_train)
m = model.coef_[0]     # eğim
b = model.intercept_   # kesişim

# 2) 1980 öncesi çizgi (orijinal)
year_min_before = int(df_before_1980["Yil"].min())
years_before = np.linspace(year_min_before, 1979, 200).reshape(-1, 1)
y_pred_before = model.predict(years_before)

# 3) 1980 sonrası çizgi: aynı eğim m, fakat 1980'de
# y(1980) = (1980 öncesi çizginin BAŞLANGIÇ y'si) olacak şekilde dikey kaydır
y_start_before = (m * year_min_before) + b          # "80 öncesi çizginin başladığı nokta"nın y'si
b_after = y_start_before - m * 1980                 # yeni kesişim
years_after = np.linspace(1980, 2030, 200).reshape(-1, 1)
y_pred_after = (m * years_after) + b_after

# 4) Çiz
plt.figure(figsize=(12,6))
plt.scatter(df["Yil"], df["Boylam"], color="gray", alpha=0.4, label="Gerçek Veriler")

# 1980 öncesi (orijinal fit)
plt.plot(years_before, y_pred_before, color="red", linewidth=2.5, label="Linear Regression (1980 Öncesi)")

# 1980 sonrası (AYNI eğim, 1980'de başlangıç y'si = 80 öncesi çizginin başlangıcı)
plt.plot(years_after, y_pred_after, color="red", linestyle="--", linewidth=2.5,
         label="Aynı Eğim, 1980'de Başlangıç = 80 Öncesi Başlangıç")

plt.axvline(1980, color="blue", linestyle="--", alpha=0.7)
plt.title("Boylam Trendleri – 80 Öncesi Çizgiyle Aynı Eğim, 80 Sonrasında Aynı Başlangıç Y'si")
plt.xlabel("Yıl"); plt.ylabel("Boylam (Longitude)")
plt.grid(True); plt.legend(); plt.tight_layout(); plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# 1980 öncesi ve sonrası böl
df_before_1980 = df[df["Yil"] < 1980].dropna(subset=["Yil", "Boylam", "xM"]).copy()
df_after_1980  = df[df["Yil"] >= 1980].dropna(subset=["Yil", "Boylam", "xM"]).copy()

# --- 1) 1980 öncesi ağırlıklı fit ---
Xb = df_before_1980[["Yil"]].values
yb = df_before_1980["Boylam"].values
wb = df_before_1980["xM"].values ** 2   # büyüklüğün karesiyle ağırlıklandır

model_before = LinearRegression().fit(Xb, yb, sample_weight=wb)
m1, b1 = model_before.coef_[0], model_before.intercept_

year_min_before = int(df_before_1980["Yil"].min())
years_before = np.linspace(year_min_before, 1979, 200).reshape(-1, 1)
y_before_line = model_before.predict(years_before)

# --- 2) Kopya trend (aynı eğim, 1980'de başlangıç y'si = 80 öncesi başlangıcı) ---
y_start_before = m1 * year_min_before + b1
b_copy = y_start_before - m1 * 1980
years_after = np.linspace(1980, 2030, 300).reshape(-1, 1)
y_after_copy = m1 * years_after + b_copy

# --- 3) 1980 sonrası ağırlıklı fit ---
Xa = df_after_1980[["Yil"]].values
ya = df_after_1980["Boylam"].values
wa = df_after_1980["xM"].values ** 2

model_after = LinearRegression().fit(Xa, ya, sample_weight=wa)
m2, b2 = model_after.coef_[0], model_after.intercept_
y_after_true = model_after.predict(years_after)
r2_after = r2_score(ya, model_after.predict(Xa), sample_weight=wa)

# --- 4) Görselleştirme ---
plt.figure(figsize=(12,6))
plt.scatter(df_before_1980["Yil"], df_before_1980["Boylam"],
            s=df_before_1980["xM"]*8, alpha=0.5, color="gray", label="Gerçek Veriler (1980 Öncesi)")
plt.scatter(df_after_1980["Yil"], df_after_1980["Boylam"],
            s=df_after_1980["xM"]*8, alpha=0.5, color="darkgray", label="Gerçek Veriler (1980 Sonrası)")

plt.plot(years_before, y_before_line, color="red", lw=2.5, label="Ağırlıklı Trend (1980 Öncesi)")
plt.plot(years_after,  y_after_copy,  color="red", ls="--", lw=2.5, label="Kopya Trend (1980+)")
plt.plot(years_after,  y_after_true,  color="green", lw=2.5, label=f"Ağırlıklı Gerçek Trend (1980+), R²={r2_after:.3f}")

plt.axvline(1980, color="blue", ls="--", alpha=0.7)

txt = (f"Ağırlıklı fit (xM²)\n"
       f"m1={m1:.5f}, b1={b1:.2f}\n"
       f"Kopya: y = {m1:.5f}·x + {b_copy:.2f}\n"
       f"m2={m2:.5f}, b2={b2:.2f}")
plt.text(0.01, 0.02, txt, transform=plt.gca().transAxes,
         fontsize=10, ha="left", va="bottom",
         bbox=dict(boxstyle="round,pad=0.4", facecolor="white", alpha=0.8, edgecolor="0.7"))

plt.title("Ağırlıklı Boylam Trendleri – Büyük Depremler Daha Etkili (xM² Ağırlıklı)")
plt.xlabel("Yıl")
plt.ylabel("Boylam (Longitude)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
from sklearn.svm import SVR
from sklearn.metrics import r2_score

# 1️⃣ Veriyi böl
df_before_1980 = df[df["Yil"] < 1980].dropna(subset=["Yil", "Boylam"]).copy()
df_after_1980  = df[df["Yil"] >= 1980].dropna(subset=["Yil", "Boylam"]).copy()

Xb = df_before_1980[["Yil"]].values
yb = df_before_1980["Boylam"].values
Xa = df_after_1980[["Yil"]].values
ya = df_after_1980["Boylam"].values

# 2️⃣ 1980 öncesi modeli eğit
model_b = SVR(kernel="rbf", C=100, epsilon=0.01, gamma="scale")
model_b.fit(Xb, yb)

# 3️⃣ 1980 öncesi çizgi
year_min_before = int(df_before_1980["Yil"].min())
year_max_before = 1979
years_before = np.linspace(year_min_before, year_max_before, 200).reshape(-1, 1)
y_before = model_b.predict(years_before)

# 4️⃣ Aynı çizgiyi 1980 sonrası tarafa kopyala
# x koordinatlarını 1980'e göre kaydırıyoruz
duration = year_max_before - year_min_before  # örn. 50 yıl
years_after_copy = years_before + (1980 - year_min_before)
y_after_copy = y_before.copy()  # birebir aynı şekil, sadece x kaydı farklı

# 5️⃣ 1980 sonrası GERÇEK trend
model_a = SVR(kernel="rbf", C=100, epsilon=0.01, gamma="scale")
model_a.fit(Xa, ya)
years_after = np.linspace(1980, 2030, 200).reshape(-1, 1)
y_after_true = model_a.predict(years_after)
r2_after = r2_score(ya, model_a.predict(Xa))

# 6️⃣ Çizim
plt.figure(figsize=(12,6))
plt.scatter(df_before_1980["Yil"], df_before_1980["Boylam"], s=25, color="gray", alpha=0.5, label="Gerçek (1980 Öncesi)")
plt.scatter(df_after_1980["Yil"],  df_after_1980["Boylam"],  s=25, color="darkgray", alpha=0.5, label="Gerçek (1980 Sonrası)")

plt.plot(years_before, y_before, color="red", lw=2.2, label="SVR(RBF) Trend (1980 Öncesi)")
plt.plot(years_after_copy, y_after_copy, color="red", ls="--", lw=2.2, label="Kopya Trend (Aynı Şekil 1980+)")
plt.plot(years_after, y_after_true, color="green", lw=2.2, label=f"Gerçek Trend (1980+), R²={r2_after:.3f}")

plt.axvline(1980, color="blue", ls="--", alpha=0.7)
plt.title("1980 Öncesi Trendin Aynısı 1980 Sonrasına Kopyalandı — SVR(RBF)")
plt.xlabel("Yıl"); plt.ylabel("Boylam (Longitude)")
plt.legend(); plt.grid(True); plt.tight_layout(); plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor

# 2) 1980 öncesi RF modeli ve trend eğrisi
rf_before = RandomForestRegressor(n_estimators=300, random_state=42)
rf_before.fit(Xb, yb)

year_min_before = int(df_before_1980["Yil"].min())
year_max_before = 1979
years_before = np.linspace(year_min_before, year_max_before, 200).reshape(-1, 1)
y_before = rf_before.predict(years_before)

# 3) Aynı eğriyi sağ tarafa x-kaydırmasıyla kopyala
# (year_min_before → 1980 olacak şekilde)
shift = 1980 - year_min_before
years_after_copy = years_before + shift
y_after_copy = y_before.copy()  # ŞEKİL BİREBİR AYNI, sadece x farklı

# 4) 1980 sonrası gerçek RF trendi
rf_after = RandomForestRegressor(n_estimators=300, random_state=42)
rf_after.fit(Xa, ya)
years_after_true = np.linspace(1980, max(2030, int(df_after_1980["Yil"].max())), 300).reshape(-1, 1)
y_after_true = rf_after.predict(years_after_true)
r2_after = r2_score(ya, rf_after.predict(Xa))

# 5) Çizim
plt.figure(figsize=(12,6))
plt.scatter(df_before_1980["Yil"], df_before_1980["Boylam"], s=25, color="gray", alpha=0.5, label="Veri (1980 Öncesi)")
plt.scatter(df_after_1980["Yil"],  df_after_1980["Boylam"],  s=25, color="darkgray", alpha=0.5, label="Veri (1980 Sonrası)")

plt.plot(years_before, y_before, color="red", lw=2.3, label="RF Trend (1980 Öncesi)")
plt.plot(years_after_copy, y_after_copy, color="red", ls="--", lw=2.3, label="Kopya Trend (Aynı Şekil 1980+)")
plt.plot(years_after_true, y_after_true, color="green", lw=2.3, label=f"Gerçek Trend (1980+), R²={r2_after:.3f}")

plt.axvline(1980, color="blue", ls="--", alpha=0.7)
plt.title("Random Forest — 1980 Öncesi Trendin Aynısı 1980 Sonrasına Kopyalandı")
plt.xlabel("Yıl"); plt.ylabel("Boylam (Longitude)")
plt.grid(True); plt.legend(); plt.tight_layout(); plt.show()
```

```python
from sklearn.ensemble import GradientBoostingRegressor

# 2) 1980 öncesi GB modeli ve trend eğrisi
gb_before = GradientBoostingRegressor(random_state=42)
gb_before.fit(Xb, yb)

year_min_before = int(df_before_1980["Yil"].min())
year_max_before = 1979
years_before = np.linspace(year_min_before, year_max_before, 200).reshape(-1, 1)
y_before = gb_before.predict(years_before)

# 3) Aynı eğriyi sağ tarafa x-kaydırmasıyla kopyala
shift = 1980 - year_min_before
years_after_copy = years_before + shift
y_after_copy = y_before.copy()  # ŞEKİL BİREBİR AYNI

# 4) 1980 sonrası gerçek GB trendi
gb_after = GradientBoostingRegressor(random_state=42)
gb_after.fit(Xa, ya)
years_after_true = np.linspace(1980, max(2030, int(df_after_1980["Yil"].max())), 300).reshape(-1, 1)
y_after_true = gb_after.predict(years_after_true)
r2_after = r2_score(ya, gb_after.predict(Xa))

# 5) Çizim
plt.figure(figsize=(12,6))
plt.scatter(df_before_1980["Yil"], df_before_1980["Boylam"], s=25, color="gray", alpha=0.5, label="Veri (1980 Öncesi)")
plt.scatter(df_after_1980["Yil"],  df_after_1980["Boylam"],  s=25, color="darkgray", alpha=0.5, label="Veri (1980 Sonrası)")

plt.plot(years_before, y_before, color="red", lw=2.3, label="GB Trend (1980 Öncesi)")
plt.plot(years_after_copy, y_after_copy, color="red", ls="--", lw=2.3, label="Kopya Trend (Aynı Şekil 1980+)")
plt.plot(years_after_true, y_after_true, color="green", lw=2.3, label=f"Gerçek Trend (1980+), R²={r2_after:.3f}")

plt.axvline(1980, color="blue", ls="--", alpha=0.7)
plt.title("Gradient Boosting — 1980 Öncesi Trendin Aynısı 1980 Sonrasına Kopyalandı")
plt.xlabel("Yıl"); plt.ylabel("Boylam (Longitude)")
plt.grid(True); plt.legend(); plt.tight_layout(); plt.show()
```

```python
from scipy.stats import pearsonr

# --- Ortak zaman aralıkları ---
year_min_before = int(df_before_1980["Yil"].min())
year_max_before = 1979
years_before = np.linspace(year_min_before, year_max_before, 200).reshape(-1, 1)
years_after  = np.linspace(1980, 2030, 200).reshape(-1, 1)

# --- Model kümesi ---
models = {
    "LinearRegression": LinearRegression(),
    "SVR(RBF)": SVR(kernel="rbf", C=100, epsilon=0.01, gamma="scale"),
    "RandomForest": RandomForestRegressor(n_estimators=300, random_state=42),
    "GradientBoosting": GradientBoostingRegressor(random_state=42)
}

results = []

for name, model in models.items():
    # 1) 1980 öncesi fit ve trend
    model.fit(Xb, yb)
    y_before = model.predict(years_before)

    # 2) Aynı şekli sağa kaydır (copy trend)
    shift = 1980 - year_min_before
    years_after_copy = years_before + shift
    y_after_copy = y_before.copy()

    # 3) 1980 sonrası fit ve trend
    model_after = type(model)(**getattr(model, 'get_params')()) if hasattr(model, 'get_params') else model
    model_after.fit(Xa, ya)
    y_after_true = model_after.predict(years_after)

    # 4) Ortak eksen oluştur (aynı uzunlukta karşılaştırmak için)
    min_len = min(len(y_after_true), len(y_after_copy))
    y_copy_aligned = y_after_copy[:min_len]
    y_true_aligned = y_after_true[:min_len]

    # 5) Korelasyon ve benzerlik yüzdesi
    corr, _ = pearsonr(y_copy_aligned, y_true_aligned)
    overlap = round(max(0, corr) * 100, 2)  # negatifse 0 kabul

    results.append({
        "Model": name,
        "Korelasyon (r)": round(corr, 3),
        "Örtüşme (%)": overlap
    })

# --- Sonuç tablosu ---
results_df = pd.DataFrame(results)
print(results_df)
```

```python
# Cell — Olasılıksal model tahminlerinden birleşik olasılık dağılımı harita üzerinde
from scipy.stats import norm

# 1) Model tahminlerini al
preds = results_df["Tahmin"].astype(float).values
model_names = results_df["Model"].tolist()

# 2) Her model için dağılım parametresi (σ = varsayım)
# Eğer model sayısı fazla ve residual'lar varsa, bunlardan sigma hesaplanabilir.
sigma = 0.25  # 0.25 derece belirsizlik (~20-25 km)

# 3) Ortak grid (boylam ekseni)
lon_min, lon_max = 25, 45
lon_grid = np.linspace(lon_min, lon_max, 800)

# 4) Her modelin PDF'i
pdfs = []
for mu in preds:
    pdfs.append(norm.pdf(lon_grid, loc=mu, scale=sigma))
pdfs = np.array(pdfs)

# 5) Ensemble PDF (normalize edilmiş toplam)
ensemble_pdf = pdfs.sum(axis=0)
ensemble_pdf /= ensemble_pdf.max()

# 6) İstatistikler
ens_mean = float(np.sum(lon_grid * ensemble_pdf) / np.sum(ensemble_pdf))
cdf = np.cumsum((ensemble_pdf[:-1] + ensemble_pdf[1:]) / 2 * (lon_grid[1] - lon_grid[0]))
cdf = np.concatenate([[0.0], cdf])
cdf /= cdf[-1]
def qcut(q): return float(np.interp(q, cdf, lon_grid))
q05, q25, q50, q75, q95 = [qcut(q) for q in (0.05, 0.25, 0.5, 0.75, 0.95)]

print(f"Probabilistic Ensemble Mean: {ens_mean:.3f}°E")
print(f"50% CI: [{q25:.3f}, {q75:.3f}]°E")
print(f"90% CI: [{q05:.3f}, {q95:.3f}]°E")

# 7) Harita üzerinde göster
plt.figure(figsize=(11,8))
m = Basemap(projection="merc", llcrnrlon=lon_min, urcrnrlon=lon_max,
            llcrnrlat=35, urcrnrlat=43, resolution='i')
m.drawcoastlines(linewidth=0.5); m.drawcountries(linewidth=0.5)
m.drawparallels(np.arange(35, 44, 2), labels=[1,0,0,0], fontsize=9)
m.drawmeridians(np.arange(int(lon_min), int(lon_max)+1, 2), labels=[0,0,0,1], fontsize=9)
m.drawmapboundary(fill_color="#DFF6FF"); m.fillcontinents(color="#f2efe9", lake_color="#DFF6FF")

# KDE bandı
y_band = true_lat
x_pts, y_pts = m(lon_grid, np.full_like(lon_grid, y_band))
plt.scatter(x_pts, y_pts, s=10 + 120*ensemble_pdf, c=ensemble_pdf, cmap="magma",
            alpha=np.clip(ensemble_pdf, 0.15, 1.0), marker='s', edgecolors='none', zorder=4)

# Gerçek nokta
x_true, y_true = m(true_lon, true_lat)
plt.scatter(x_true, y_true, color="red", s=120, edgecolors="black", zorder=6, label="Gerçek Silivri")

# Ortalama ve güven aralıkları
x_mean, y_mean = m(ens_mean, y_band)
plt.scatter(x_mean, y_mean, s=160, marker="*", color="gold", edgecolors="black", linewidths=1.0,
            zorder=7, label=f"Ensemble Prob. Mean ({ens_mean:.2f}°E)")

for lon_ci, style, col, lab in [
    (q25, "--", "k", "50% CI"),
    (q75, "--", "k", None),
    (q05, ":",  "k", "90% CI"),
    (q95, ":",  "k", None),
]:
    x_ci1, y_ci1 = m(lon_ci, y_band - 0.4)
    x_ci2, y_ci2 = m(lon_ci, y_band + 0.4)
    plt.plot([x_ci1, x_ci2], [y_ci1, y_ci2], style, color=col, lw=1.6, zorder=6, label=lab)

plt.title("Silivri — Olasılıksal Ensemble Dağılımı (Model Tahminlerinden)")
plt.legend(loc="lower right", fontsize=9)
plt.tight_layout()
plt.show()
```

# Büyüklük - Süre ilişkisi


```python
from scipy.stats import pearsonr, spearmanr

# --- 1) Tarih sütununu datetime formatına çevir (örnek: 'Tarih' veya 'Date')
# Eğer zaten datetime ise bu satır hata vermez.
df["Tarih"] = pd.to_datetime(df["Tarih"], errors="coerce")

# --- 2) Tarihe göre sırala
df_sorted = df.sort_values("Tarih").reset_index(drop=True)

# --- 3) Bir önceki depremden geçen süreyi (gün cinsinden) hesapla
df_sorted["Delta_t_days"] = df_sorted["Tarih"].diff().dt.total_seconds() / (3600 * 24)

# İlk satırın farkı NaN olur, drop edelim
df_analyzed = df_sorted.dropna(subset=["Delta_t_days", "xM"]).copy()

# --- 4) Korelasyon analizi
pearson_corr, _ = pearsonr(df_analyzed["xM"], df_analyzed["Delta_t_days"])
spearman_corr, _ = spearmanr(df_analyzed["xM"], df_analyzed["Delta_t_days"])

print(f"Pearson korelasyonu (doğrusal ilişki): {pearson_corr:.3f}")
print(f"Spearman korelasyonu (monoton ilişki): {spearman_corr:.3f}")

# --- 5) Görselleştirme (log ölçekli süre vs büyüklük)
plt.figure(figsize=(10,6))
plt.scatter(df_analyzed["Delta_t_days"], df_analyzed["xM"], alpha=0.6, color="royalblue", s=30)
plt.xscale("log")
plt.xlabel("Bir Önceki Depremden Geçen Süre (gün, log ölçek)")
plt.ylabel("Deprem Büyüklüğü (xM)")
plt.title("Deprem Büyüklüğü vs Önceki Depremden Geçen Süre")
plt.grid(True, which="both", ls="--", lw=0.5, alpha=0.6)
plt.tight_layout()
plt.show()
```
